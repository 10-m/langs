```python
import pandas as pd

df = pd.read_csv('data/user.csv')
df
```

```python
# 文字列の一括処理
df['CommentLen'] = df['Comment'].str.len()
df
```

```python
# 文字列の繰り返し処理
sliced_list = []
for row in df.itertuples():
    sliced_list.append(row.Comment[:15])

df['SlicedComment'] = sliced_list
df
```

```python
# itertuplesの代わりにiterrows
for _, row in df.iterrows():
    assert row.Comment[:15] == row.SlicedComment
```

```python
def get_sliced_str(x):
    return x[:15]
# 関数を使って作成
df['SlicedComment'] = df['Comment'].apply(get_sliced_str)
df
```

```python
# strアクセサを使って作成
df['SlicedComment'] = df['Comment'].str[:15]
df
```
