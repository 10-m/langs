```python
%matplotlib inline
import pandas as pd

df = pd.read_table('data/attendances.csv')
df.head()
```

```python
df.plot();
```

```python
# x軸、y軸の列を指定
df.plot(x="Date", y='Office', title="Attendances");
```

```python
# Seriesのグラフ
series = df["Tele"]
series.head()
```

```python
series.plot();
```

```python
# 棒グラフ
df.plot.bar(x="Date");
```

```python
df.plot.bar(x="Date", stacked=True);
```

```python
# 散布図
from datetime import datetime
def get_day_of_week(str_date):
    date = datetime.strptime(str_date,'%Y-%m-%d')
    return date.weekday()

df["day_of_week"] = df["Date"].apply(get_day_of_week)
df.head()
```

```python
df.plot.scatter(x="day_of_week", y="Office");
```

```python
# 箱ひげ図
df.plot.box(showmeans=True);
```

```python
# ヒストグラムの確認
df.hist(bins=10, range=(0, 100));
```

```python
# 散布図行列
pd.plotting.scatter_matrix(df);
```

```python
# 相関行列
df.corr()
```
